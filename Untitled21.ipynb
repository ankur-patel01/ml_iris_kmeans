<h4 style="color:green"> <i>Importing libraries</i></h4>

```python
#importing all libraries required for this notebook
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

<h4 style="color:green"> <i>Reading Data</i></h4>

```python
#reading a csv file
data_set=pd.read_csv("Iris.csv")
print("The data has been imported, as shown below")
data_set.head()
```

<h4 style="color:green"> <i>Encoding the Categorical Data</i></h4>

```python
#we will encode the species column
ds=data_set['Species']
obj=LabelEncoder()
data_set['Spe_enc']=obj.fit_transform(ds)
print("we have added a column named 'spe_enc' as it is encoded data")
print((data_set.iloc[:,-1]).values)
```

<h4 style="color:green"> <i>Visualising the Actual Data</i></h4>

```python
x1=data_set.iloc[:,[0,1,2,3,4]].values
y1=data_set.iloc[:,-1].values
```

```python
#plotting the distribution of Original Iris Data
plt.scatter(x1[y1==0,0],x1[y1==0,1],label = 'Iris-setosa')
plt.scatter(x1[y1==1,0],x1[y1==1,1],label = 'Iris-versicolour')
plt.scatter(x1[y1==2,0],x1[y1==2,1],label = 'Iris-virginica')
style.use('ggplot')
plt.title("CLUSTERS OF ORIGINAL IRIS DATA")
plt.legend()
plt.show()
```

### as we can see from the above graph that the relation in data.

<h4 style="color:green"> <i>Finding the optimum number of clusters for k-means classification from elbow plot method.
</i></h4>

```python
# the elbow method uses the WCSS (Within-Cluster-Sum-Squares) concept to draw the plot
# by plotting WCSS values on the Y-axis and
# the number of clusters on the X-axis.
wcc=[]
for i in range(1,11):
    kms=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kms.fit(x1)
    wcc.append(kms.inertia_)
plt.title('Elbow Method Plot')
plt.plot(range(1, 11), wcc,label='Clusters linear')
plt.legend()
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

<h4 style="color:green"> <i>Visualizing the actual data
</i></h4>

```python
plt.plot(range(1, 11), wcc,label='Clusters linear')
plt.scatter(3,wcc[2],label='Elbow Point')
plt.title('Elbow Point')
plt.legend()
plt.xlabel('No. of Clusters')
plt.ylabel('Within Cluster Sum of Squares (WCSS)')
plt.show()
```

<h3 style="color:green"> <i>Building Model</i></h3>
<h4 style="color:green"> <i>Defining and Fitting Model</i></h4>
<h4 style="color:green"> <i>Training the K-means algorithm on the dataset</i></h4>

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kms=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_predict=kms.fit_predict(x1)
```

<h4 style="color:green"> <i>Visualizing the Clusters</i></h4>

```python
#plotting the distribution of Predicted Data
plt.scatter(x1[y_predict==0,0],x1[y_predict==0,1],label = 'Iris-setosa')
plt.scatter(x1[y_predict==1,0],x1[y_predict==1,1],label = 'Iris-versicolour')
plt.scatter(x1[y_predict==2,0],x1[y_predict==2,1],label = 'Iris-virginica')
plt.scatter(kms.cluster_centers_[:,0],kms.cluster_centers_[:,1],label = 'Centroids',c='yellow')
style.use('ggplot')
plt.title("CLUSTERS AFTER PREDICTION")
plt.legend()
plt.show()
```
